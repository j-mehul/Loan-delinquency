```python
y=df.m13 
train = df.drop('m13',1)
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y= train_test_split(train,y, test_size=0.20, random_state=2019,stratify = y)
```

```python
import xgboost as xgb
feature=train_x.columns.tolist()
```

```python
param = {'objective' : 'binary:logistic',
            'eta' : 0.01,
            'max_depth' : 6,
            'silent' : 1,
            'eval_metric' : "auc",
             'booster' : 'gbtree',
            'min_child_weight' : 5,
            'nthread' : 16,
            'subsample' : 0.9,
            'colsample_bytree' : 0.7,
            'class_weight':'balanced',
            'seed' : 123}

para = list(param.items())

xgtrain = xgb.DMatrix(train_x, train_y, feature_names=feature)
xgval = xgb.DMatrix(val_x, val_y, feature_names=feature)
#xgtest = xgb.DMatrix(test, feature_names=feature)
    
watchlist = [ (xgtrain,'train'), (xgval, 'valid') ]
model = xgb.train(para, xgtrain, num_boost_round=3000, evals= watchlist, early_stopping_rounds=20, verbose_eval = 10)   
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 7))
xgb.plot_importance(model, max_num_features=30, height=0.5, ax=ax)
```
